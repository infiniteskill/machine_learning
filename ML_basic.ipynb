```python
#Classification Matrices>> Either Model is correct on prediction or its not , matrices have same idea
## 1. Accuracy  >> Number of correct predictions/total number of predictions
## accuracy is useful when target classes are well balanced , means same number of images/data for each  class

## 2. Recall >> Ability of model to find all relevant cases within a dataset, 
## Recall = (number of true positive)/(number of true positive + number of false negative)
## recall expresses the ability to find all relevant instances in dataset

## 3. Precision = Number of true positive/(number of ture positive+number of false positive) 
## precision expresses the proprtion of the datapoints our model says was relevant were actually relevant


## 4. F1-Score >> harmonic mean of precison of recall
## F1 =2*(precision*recall)/(precision+recall)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('fake_reg.csv')
```

```python
df
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['feature1','feature2']].values
```

```python
y=df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(  X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
# model=Sequential([Dense(4,activation='relu'),Dense(2,activation='relu'),Dense(1)])
```

```python
model=Sequential()
model.add(Dense(8,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(2,activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='rmsprop', loss='mse')
```

```python
model.fit(x=X_train,y=y_train,epochs=1500)   # x is training data , y is label
```

```python
loss_df=pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_predictions=model.predict(X_test)
```

```python
test_predictions=pd.Series(test_predictions.reshape(300))
```

```python
pred_df=pd.DataFrame(y_test,columns=['Test True y'])
```

```python
pred_df=pd.concat([pred_df,test_predictions],axis=1)
```

```python
pred_df
```

```python
sns.scatterplot(data=pred_df)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(pred_df['Test True y'],pred_df[0])
```
